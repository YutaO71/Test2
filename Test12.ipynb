```python
def analyze_student_performance(grades):
    weak_modules = []
    total = 0
    
    for module, score in grades.items():
        total += score
        if score < 60:
                weak_modules.append(module)
    average = total / len(grades) if grades else 0
    return average, weak_modules
sample_grades = {'Math': 70, 'Science': 55, 'English': 80}
avg, weak = analyze_student_performance(sample_grades)
print(f"Average: {avg}")
print(f"Weak modules: {weak}")
%history -f Test2.py
```
